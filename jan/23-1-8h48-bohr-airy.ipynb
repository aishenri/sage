La meilleure façon de comprendre la perméabilité/permitivité est de se les représenter comme une limite entre deux intervales. Un cercle inscrit dans un cercle circonscrit :

```python
a=360E-8;b=360E+8;L=1/n(pi)*a;C=1/(n(pi)*b);ç=1/sqrt(L*C)
show(LatexExpr(r"\large \mu=L=\frac{1}{\pi a}="),L)
show(LatexExpr(r"\large \epsilon=C=\frac{1}{\pi b}="),C)
show(LatexExpr(r"\large ç=\frac{1}{\sqrt L C}="),ç)
show(LatexExpr(r"\large c=\frac{1}{\sqrt \mu \epsilon}=299792458="),c)
```
