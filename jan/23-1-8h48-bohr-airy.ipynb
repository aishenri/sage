La meilleure faon de comprendre la permabilit/permitivit est de se les reprsenter comme une limite entre deux intervales. Un cercle inscrit dans un cercle circonscrit :

```python
a=360E-8;b=360E+8;L=1/n(pi)*a;C=1/(n(pi)*b);=1/sqrt(L*C)
show(LatexExpr(r"\large \mu=L=\frac{1}{\pi a}="),L)
show(LatexExpr(r"\large \epsilon=C=\frac{1}{\pi b}="),C)
show(LatexExpr(r"\large =\frac{1}{\sqrt L C}="),)
show(LatexExpr(r"\large c=\frac{1}{\sqrt \mu \epsilon}=299792458="),c)
```
